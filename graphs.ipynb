```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_no_vae= pd.read_csv("training_no_vae.log")
df_vae = pd.read_csv("training_vae.log")
```

```python
fig = plt.figure(figsize=(15,10))

plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")

plt.plot(df_no_vae['epoch'], df_no_vae['loss'], label='Training Loss (No VAE)')
plt.plot(df_no_vae['epoch'], df_no_vae['val_loss'], linestyle='--', label='Validation Loss (No VAE)')
plt.plot(df_vae['epoch'], df_vae['loss'], label='Training Loss')
plt.plot(df_vae['epoch'], df_vae['val_loss'], linestyle='--', label='Validation Loss')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(15,10))

plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.ylim(0, 1)

plt.plot(df_no_vae['epoch'], df_no_vae['acc'], label='Training Accuracy (No VAE)')
plt.plot(df_no_vae['epoch'], df_no_vae['val_acc'], linestyle='--', label='Validation Accuracy (No VAE)')
plt.plot(df_vae['epoch'], df_vae['acc'], label='Training Accuracy')
plt.plot(df_vae['epoch'], df_vae['val_acc'], linestyle='--', label='Validation Accuracy')
plt.legend()
plt.show()
```
