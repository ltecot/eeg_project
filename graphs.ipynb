```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_no_vae= pd.read_csv("training_no_vae.log")
df_vae = pd.read_csv("training_vae.log")
```

```python
fig = plt.figure(figsize=(6,5))

plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")

size = min(len(df_no_vae['epoch']), len(df_vae['epoch']))

plt.plot(df_no_vae['epoch'][:size], df_no_vae['loss'][:size], label='Training Loss (No VAE)', color='red')
plt.plot(df_no_vae['epoch'][:size], df_no_vae['val_loss'][:size], linestyle='--', label='Validation Loss (No VAE)', color='orange')
plt.plot(df_vae['epoch'][:size], df_vae['loss'][:size], label='Training Loss', color='blue')
plt.plot(df_vae['epoch'][:size], df_vae['val_loss'][:size], linestyle='--', label='Validation Loss', color='purple')
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(6,5))

plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.ylim(0, 1)

plt.plot(df_no_vae['epoch'][:size], df_no_vae['acc'][:size], label='Training Accuracy (No VAE)', color='red')
plt.plot(df_no_vae['epoch'][:size], df_no_vae['val_acc'][:size], linestyle='--', label='Validation Accuracy (No VAE)', color='orange')
plt.plot(df_vae['epoch'][:size], df_vae['acc'][:size], label='Training Accuracy', color='blue')
plt.plot(df_vae['epoch'][:size], df_vae['val_acc'][:size], linestyle='--', label='Validation Accuracy', color='purple')
plt.legend()
plt.show()
```
